```python
import pandas as pd
```

```python
mf = pd.read_csv('/content/drive/MyDrive/data maining/MITM ARP Spoofing.csv')
```

```python
mf
```

```python
rf = pd.read_csv('/content/drive/MyDrive/data maining/MQTT Malformed.csv')
```

```python
rf
```

```python
wr = pd.read_csv('/content/drive/MyDrive/data maining/DoS UDP Flood.csv')
```

```python
wr
```

```python
hasilgabung =pd.concat([mf,rf,wr],ignore_index=True)
```

```python
hasilgabung.columns.values
```

```python

A =  hasilgabung.iloc[:,7:76]
```

```python
A
```

```python
T = hasilgabung.iloc[:,83:84]
```

```python
T
```

```python
from sklearn.model_selection import train_test_split
```

```python
A_train,A_test,T_train,T_test = train_test_split(A,T, test_size = 0.2, random_state = 42)
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
```

```python
laras = DecisionTreeClassifier(criterion='entropy', splitter = 'random')
laras.fit(A_train,T_train)
```

```python
T_pred = laras.predict(A_test)
```

```python
T_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(T_test,T_pred)
```

```python
accuracy
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure(figsize = (10, 7))
tree.plot_tree(laras, feature_names = A.columns.values, class_names = np.array([ 'MITM ARP Spoofing','DoS UDP Flood','MQTT Malformed']), filled = True)
plt.show()
```

```python
import seaborn as wow
from sklearn import metrics
label = np.array([ 'MITM ARP Spoofing','DoS UDP Flood','MQTT Malformed'])
```

```python
import matplotlib.pyplot as plt
conf_matrix = metrics.confusion_matrix(T_test, T_pred)
plt.figure(figsize=(10, 10))
wow.heatmap(conf_matrix, annot=True, xticklabels=label, yticklabels=label)
plt.xlabel('Prediksi')
plt.ylabel('Jaya jaya')
plt.show()
```
